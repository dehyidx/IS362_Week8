# Cleaned "Auto MPG" data set

```python
import numpy as np
import pandas as pd

import seaborn as sns
# matplotlib
import matplotlib as mpl
# matplotlib plotting functions

import matplotlib.pyplot as plt
# we want our plots inline
%matplotlib inline 
#import seaborn as sns




auto = pd.read_fwf(r'C:\Users\ghydi\Downloads\auto+mpg\auto-mpg.data',
                  names=['mpg', 'cylinders', 'displacement',
                         'horsepower', 'weight', 'acceleration', 'model year', 
                         'origin', 'car name'])

auto.loc[auto["origin"] == 1, 'origin'] = 'USA'
auto.loc[auto["origin"] == 2, 'origin'] = 'Asia'
auto.loc[auto["origin"] == 3, 'origin'] = 'Europe'

auto.loc[auto['horsepower'] == '?', 'horsepower'] = np.nan
auto['horsepower'] = pd.to_numeric(auto['horsepower'])

auto
```

# Cylinders distribution chart

```python
auto.cylinders.hist()
```

# Horsepower & weight scatterplot

```python
auto.plot(kind='scatter', x='horsepower', y='weight')
```

# Question: Is there a correlation between weight and acceleration?

```python
sns.regplot(x='weight', y='acceleration', data=auto)
```

As shown by the graph above, there is a slight negative correlation between weight and horsepower, but there are too many outliers for it to be definitive.
